```python
import random

words = ["apple", "banana", "orange", "grape", "lemon"]

hangman_pics = [
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---""",
        """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
]

class hangmangame:
    def __init__(self):
        self.word = random.choice(words)
        if len(self.word) == 5:
            self.guess = ['_','_','_','_','_',]
        else:
            self.guess = ['_','_','_','_','_','_']
        self.count = 0
        self.comp_word = list(self.word)
    def input_letter(self):
        self.letter = input('하나의 철자를 입력해주세요 :  ')

    def guess_state(self):   #어디까지 추측했는지 볼 수 있도록 출력해주는 함수
        for i in range(len(self.guess)):
            print(self.guess[i],end = ' ')

    def add_letter(self):
        for i in range(len(self.comp_word)):
            if self.letter == self.comp_word[i]:
                self.guess[i] = self.letter

    def play_hangman(self):
        while self.count < 6:
            print(hangman_pics[6-self.count])

            self.guess_state()
            self.input_letter()

            if self.letter in self.guess:
                print("이미 추측한 글자입니다.")
                self.count += 1
                print(f"남은 시도 횟수 :{6-self.count}" )
            elif self.letter in self.comp_word:
                self.add_letter()
                self.guess_state()
                print("추측에 성공하셨습니다.")

                if self.guess == self.comp_word:
                    print("축하합니다! 단어를 맞추셨습니다. 시도횟수:",self.count)
                    break
                else :
                    self.count += 1
                    print(f"남은 시도 횟수 :{6-self.count}" )

            else:
                self.count +=1
                print(f"틀렸습니다. 남은 시도 횟수 :{6-self.count}" )

            if self.count == 6:
                print(hangman_pics[0])
                print("남은 시도 횟수가 없습니다.")
                print(f"게임오버!\n정답:{self.word}")

gamestart = hangmangame()
gamestart.play_hangman()
```
